# Notebook 1 — Préprocessing
Chargement et nettoyage du dataset Twitter Bot Detection (Zenodo 5574403).
8 386 comptes labellisés : 0 = humain, 1 = bot.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../data/twitter_bots.csv')
print('Shape:', df.shape)
df.head()
```

```python
print('Types de colonnes:')
print(df.dtypes.value_counts())
print('\nValeurs manquantes:', df.isnull().sum().sum())
```

```python
print('Distribution des labels:')
print(df['class_bot'].value_counts())
print(f'Taux de bots : {df["class_bot"].mean()*100:.1f}%')
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

df['class_bot'].value_counts().plot.bar(ax=axes[0], color=['steelblue', 'crimson'])
axes[0].set_xticklabels(['Humain', 'Bot'], rotation=0)
axes[0].set_title('Distribution des classes')
axes[0].set_ylabel('Nombre de comptes')

df.boxplot(column='followers_count', by='class_bot', ax=axes[1])
axes[1].set_title('Followers : humains vs bots')
axes[1].set_xlabel('0 = Humain, 1 = Bot')
plt.suptitle('')
plt.tight_layout()
plt.savefig('../plots/preprocessing_overview.png', dpi=120)
plt.show()
```

```python
df_clean = df.fillna(0)
df_clean.to_parquet('../data/twitter_clean.parquet', index=False)
print('Sauvegardé : data/twitter_clean.parquet', df_clean.shape)
```
