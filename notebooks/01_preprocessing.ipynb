# 01 — Préprocessing
Chargement et nettoyage du dataset Twitter Bot Detection (Zenodo 5574403).
**Source** : https://zenodo.org/records/5574403 — CC-BY 4.0, collecté sans API Twitter.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/twitter_bots.csv')
print(df.shape)
df.head()
```

## Structure des données

```python
df.info()
```

```python
print('Valeurs manquantes :', df.isnull().sum().sum())
```

## Distribution des labels

```python
print(df['class_bot'].value_counts())
print(f'Taux de bots : {df["class_bot"].mean()*100:.1f}%')
```

```python
df['class_bot'].value_counts().plot.bar(color=['steelblue', 'crimson'])
plt.xticks([0, 1], ['Humain', 'Bot'], rotation=0)
plt.title('Distribution des classes')
plt.ylabel('Nombre de comptes')
plt.tight_layout()
plt.show()
```

## Aperçu des distributions clés

```python
print('Followers — médiane humains :', df[df['class_bot']==0]['followers_count'].median())
print('Followers — médiane bots    :', df[df['class_bot']==1]['followers_count'].median())
```

```python
print('Âge du compte — médiane humains :', df[df['class_bot']==0]['account_age'].median(), 'jours')
print('Âge du compte — médiane bots    :', df[df['class_bot']==1]['account_age'].median(), 'jours')
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
for i, (col, title) in enumerate([('followers_count', 'Followers'), ('account_age', 'Âge du compte (jours)')]):
    df.groupby('class_bot')[col].median().plot.bar(ax=axes[i], color=['steelblue', 'crimson'])
    axes[i].set_xticklabels(['Humain', 'Bot'], rotation=0)
    axes[i].set_title(f'Médiane — {title}')
plt.tight_layout()
plt.savefig('../plots/preprocessing_overview.png', dpi=120)
plt.show()
```

## Nettoyage et sauvegarde

```python
df_clean = df.fillna(0)
print('Valeurs manquantes après nettoyage :', df_clean.isnull().sum().sum())
```

```python
df_clean.to_parquet('../data/twitter_clean.parquet', index=False)
print('Sauvegardé : data/twitter_clean.parquet', df_clean.shape)
```
